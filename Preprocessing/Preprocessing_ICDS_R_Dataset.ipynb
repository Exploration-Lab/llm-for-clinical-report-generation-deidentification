#including Summary information

```python
def func(txt, tgt, summ):
  words = []
  ctag  = []
  sum =[]
  j = 0
  for i in tgt:
    k = int(i['start_offset'])
    l = int(i['end_offset'])
    for tt in txt[j:k].strip().split(" "):
      tt = tt.replace("\n", '')
      tt = tt.replace("\n", '').strip()
      if tt != '':
        words.append(tt)
        ctag.append("O")
    info = i["fake_entity"].split(" ")
    ctag.append(f"B-{i['label']}")
    for _ in info[1:]:
      ctag.append(f"I-{i['label']}")
    words += [ii.strip() for ii in info]
    j = l
  for tt in txt[j:len(txt)].strip().split(" "):
    tt = tt.replace("\n", '')
    tt = tt.replace("\n", '').strip()
    if tt != '':
      words.append(tt)
      ctag.append("O")
  for i in range(len(words)):
    sum.append(summ)
  return words, ctag, sum
```

```python
def chunk_dataframe(df):
    chunks = []
    chunk_num = 1
    for _, group in df.groupby('summary #'):
        num_rows = len(group)
        num_chunks = num_rows // 100
        remainder = num_rows % 100
        for i in range(num_chunks):
            chunk = group.iloc[i * 100: (i + 1) * 100].copy()
            chunk['chunk'] = f'chunk {chunk_num} {i + 1}'
            chunks.append(chunk)
        if remainder > 0:
            chunk = group.iloc[num_chunks * 100:].copy()
            chunk['chunk'] = f'chunk {chunk_num} {num_chunks + 1}'
            chunks.append(chunk)
        chunk_num += 1
    return pd.concat(chunks)
```

```python
df_train = {'Doc #' : [], 'Words' : [], 'Tags' : [], 'segment #': [], 'summary #':[]}

for i in tqdm(range(train.shape[0])):
  txt = train.iloc[i]['text']
  # print(txt)
  tgt = train.iloc[i]['entities']
  summ = train.iloc[i]['document_index']
  a,b,c = func(txt, tgt, summ)
  # txt = train.iloc[i]['text']
  # tgt = json.loads(str(train.iloc[i]['entities']))  # Use json.loads() instead of ast.literal_eval()
  # a, b = func(txt, tgt)
  df_train["Words"] += a
  df_train["Tags"] += b
  df_train['Doc #'] += [f"Doc-{i}"]* len(a)
  df_train['summary #'] += c
  x = int(len(a)/100)
  y = len(a)%100
  for j in range(x):
    df_train['segment #'] += [f"segment {i} {j}"]*100
  df_train['segment #'] += [f"segment {i} {j+1}"]*y
# df_train = pd.DataFrame(df_train)
```
